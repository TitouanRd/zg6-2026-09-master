Dès l’Antiquité, le lien entre musique et mathématiques est établi par Pythagore (VIᵉ siècle av. J.-C.), qui montre que les intervalles musicaux correspondent à des rapports simples de longueurs de cordes. La corde vibrante devient ainsi un objet central pour comprendre le son.

Au XVIIᵉ siècle, avec la naissance de la physique moderne, Galilée (1638) et Mersenne (1636) étudient expérimentalement la hauteur des sons produits par une corde tendue, reliant fréquence, longueur, tension et masse linéique.

En 1747, d’Alembert formule pour la première fois l’équation des ondes pour une corde vibrante et en donne la solution générale. Il impose toutefois des conditions de régularité fortes sur la forme initiale de la corde.

Peu après, Euler (1748–1750) critique ces restrictions et accepte des formes initiales plus générales, introduisant une conception élargie de la notion de fonction, encore mal définie à l’époque.

En 1753, Daniel Bernoulli affirme que tout mouvement d’une corde peut être décomposé en une somme d’oscillations harmoniques, correspondant aux sons musicaux observés. Il introduit l’idée de développement en séries trigonométriques, vivement contestée par d’Alembert et Euler.

Cette controverse, à la fois mathématique, physique et philosophique, révèle l’insuffisance des outils analytiques du XVIIIᵉ siècle. Elle sera progressivement résolue au début du XIXᵉ siècle avec Fourier (1807–1822), puis Dirichlet et Cauchy, qui donnent un cadre rigoureux aux séries et à la notion de fonction.

L’équation des cordes vibrantes occupe ainsi une place fondatrice : issue de la musique antique, elle conduit à la naissance des équations aux dérivées partielles, de l’analyse moderne et de la physique mathématique des ondes.

# Introduction sur les équations aux dérivées partielles

## 0-Problème de la corde vibrante

### 0-a) Equation de d'Alembert

<p align="center">
  <img src="corde_PFD.png" width="400"/>
</p>

Hypothèses : i) poids de la corde négligée, ii) corde inextensible (se courbe mais ne s'allonge pas), infiniment souple (en flexion) et homogène de masse linéique $\mu$, iii) corde tendue horizontalement à une force constante $T(x,t)$ et excitée verticalement, iv) frottements de l'air négligés, v) déplacements supposés faibles. On suppose d'après v) que $\alpha \approx u_{,x}$.

On isole une portion de corde et on applique le théorème de la résultante dynamique :

\begin{equation}
\mu \, dx \, u_{,tt} (x,t) \overrightarrow{y} = \overrightarrow{T}(x+dx,t)-\overrightarrow{T}(x,t)
\end{equation}

On projette sur $\overrightarrow{x} $ :
\begin{equation}
0=T(x+dx,t) \cos{\alpha(x+dx,t)}-T(x,t) \cos{\alpha(x,t)} \approx T_{,x} (x,t) dx \Rightarrow T(x,t)=T_0
\end{equation}

$T(x,t)=T_0$ est constante le long de la corde et prise constante dans le temps.

On projette sur $\overrightarrow{y} $ :
\begin{equation}
\mu \, dx \, u_{,tt} (x,t)=T(x+dx,t) \sin{\alpha(x+dx,t)}-T(x,t) \sin{\alpha(x,t)} \approx T_0 \alpha_{,x} dx = T_0 u_{,xx} dx
\end{equation}

Il vient l'équation de d'Alembert :

\begin{equation}
u_{,tt}=c^2 u_{,xx}
\end{equation}

Avec $c=\sqrt{\frac{T_0}{\mu}}$ (m/s) la célérité de l'onde.

### 0-b) Forme générale de la solution

Afin de résoudre ce problème, on effectue le changement de variable suivant :

\begin{equation}
\begin{cases}
p=t-\frac{x}{c} \Rightarrow \frac{\partial p}{\partial t}=1 \quad \text{et} \quad \frac{\partial p}{\partial x}=-\frac{1}{c} \\
q=t+\frac{x}{c} \Rightarrow \frac{\partial q}{\partial t}=1 \quad \text{et} \quad \frac{\partial q}{\partial x}=+\frac{1}{c} \\
\end{cases}
\end{equation}

Considérant les anciennes variables en fonction des nouvelles, on en déduit les opérateurs différentiels suivant :

\begin{equation}
\begin{cases}
x=x(p,q) \Rightarrow \frac{\partial}{\partial x}=\frac{\partial }{\partial p} \frac{\partial p}{\partial x} + \frac{\partial }{\partial q} . \frac{\partial q}{\partial x} = \frac{1}{c} \left( -\frac{\partial}{\partial p} + \frac{\partial }{\partial q} \right) \\
t=t(p,q) \Rightarrow \frac{\partial}{\partial t}=\frac{\partial }{\partial p} \frac{\partial p}{\partial t} + \frac{\partial }{\partial q} . \frac{\partial q}{\partial t} = \frac{\partial}{\partial p} + \frac{\partial }{\partial q}
\end{cases}
\end{equation}

Ainsi, l'EDP nous permet d'écrire, sous la forme d'opérateurs :

\begin{equation}
\left( \frac{\partial^2}{\partial x^2} - \frac{1}{c^2} \frac{\partial^2}{\partial t^2} \right) u=\left( \frac{\partial}{\partial x} - \frac{1}{c} \frac{\partial}{\partial t} \right) \left( \frac{\partial}{\partial x} + \frac{1}{c} \frac{\partial}{\partial t} \right) u = 0
\end{equation}

C'est à dire,

\begin{equation}
\frac{1}{c}\left( -\frac{\partial}{\partial p} +\frac{\partial}{\partial q} - \left( \frac{\partial}{\partial p} +\frac{\partial}{\partial q} \right) \right) \left( -\frac{\partial}{\partial p} +\frac{\partial}{\partial q} + \left( \frac{\partial}{\partial p} +\frac{\partial}{\partial q} \right) \right) u = 0
\end{equation}

Et finalement, 

\begin{equation}
\left( \frac{\partial}{\partial p} \right)  \left( \frac{\partial}{\partial q} \right) u = \frac{\partial^2 u}{\partial p \partial q} = 0
\end{equation}

Après double intégration, on obtient la forme générale de la solution sous la forme d'une onde plane progressive (O.P.P.) par :

\begin{equation}
u(x,t)=f(p)+g(q) = f \left(t-\frac{x}{c} \right)+g \left(t+\frac{x}{c} \right)
\end{equation}

Avec $f$ et $g$ deux fonctions quelconques, pourvu qu'elles soient continues et deux fois dérivables.

L'interprétation physique est donnée ci après. Posons $\Delta x = c \Delta t$, où $\Delta x$ est la distance parcourue par l'onde sur une durée $\Delta t$. Il vient alors :

\begin{equation}
f \left(t-\frac{x}{c} \right) = f \left(t-\frac{x}{c} + \Delta t - \frac{\Delta x}{c} \right)=f \left(t + \Delta t-\frac{x+\Delta x}{c}\right) \Rightarrow f(x,t)=f(x+\Delta x,t+\Delta t)
\end{equation}

Le signal représenté par $f$ se propage donc identique à lui même à la vitesse $c$.

<p align="center">
  <img src="corde_propagation.png" width="400"/>
</p>

Le raisonnement est analogue pour $g$. La solution $u(x,t)= f \left(t-\frac{x}{c} \right)+g \left(t+\frac{x}{c} \right)$ représente ainsi la superposition de 2 O.P.P. se propageant à la même vitesse suivant l'axe Ox, mais en sens opposé. Cette solution générale ne tient pas compte des Conditions aux Limites (C.L.), ni des Conditions Initiales (C.I.).

### 0-c) Résolution pour une corde fixée en ses extrémités

<p align="center">
  <img src="corde_fix.png" width="400"/>
</p>

La corde est de longueur $L$ est ses extrémités sont placées en $x=0$ et $x=L$. Les conditions aux limites sont données par :

\begin{equation}
\begin{cases}
u(x=0,t)=0 \quad \forall t \\
u(x=L,t)=0 \quad \forall t
\end{cases}
\end{equation}

Avant de passer à la résolution au sens propre, on peut mettre en évidence une propriété de la solution.

Les CL appliquées à la solution générale donnent :

\begin{equation}
\forall t, \quad u(x=0,t)=f(t)+g(t)=0 \Rightarrow f(t)=-g(t) \\
\end{equation}

Ce qui donne pour solution $u(x,t)= f \left(t-\frac{x}{c} \right) -f \left(t+\frac{x}{c} \right)$. Et, 

\begin{equation}
\forall t, \quad u(x=L,t)=f\left(t-\frac{L}{c} \right) -f\left(t+\frac{L}{c} \right)=0 \Rightarrow f\left(t-\frac{L}{c} \right)=f\left(t+\frac{L}{c} \right) \\
\end{equation}

En posant $t'=t-\frac{l}{c}$, on aboutit à $f(t')=f \left(t'+2\frac{L}{c} \right)$. $u$ est donc périodique en $t$ de période $2\frac{L}{c}$ et périodique en $x$ de période $2L$. Ce résultat permet d'affirmer l'existence d'un développement en série de Fourier, comme on le verra plus loin.

Dans la suite, on suppose que la solution est de la forme $u(x,t)=X(x) T(t)$ pour permettre les conditions aux limites quelque soit $t$.

En écrivant que $X_{,xx}(x)=X''(x)$ et $T_{,tt}(t)=\ddot{T}(t)$, l'équation de d'Alembert devient :

\begin{equation}
c^2 X''(x) T(t)= X(x) \ddot{T}(t) \Rightarrow c^2\frac{X''(x)}{X(x)}= \frac{\ddot{T}(t)}{T(t)}=-\omega^2=-k^2 c^2
\end{equation}

Avec $-\omega^2$ une constante puisque le membre fonction de $x$ ne peut être égal au membre fonction de $t$, quelques soient $x$ et $t$ que s'ils sont constants. Le signe de la constante est forcément négatif. Dans le cas contraire, la solution de forme exponentielle ne permettrait pas de satisfaire les deux CL. Il vient :

\begin{equation}
\begin{cases}
X''(x)+k^2 X(x)=0 \\
\ddot{T}(t)+\omega^2 T(t)=0
\end{cases}
\quad
\Rightarrow
\quad
\begin{cases}
X(x)=A \, cos(kx+\phi) \\
T(t)=B \, cos(\omega t + \psi)
\end{cases}
\end{equation}

Finalement,

\begin{equation}
u(x,t)=C \, cos(kx+\phi) cos(\omega t + \psi)
\end{equation}

On applique les conditions aux limites :

\begin{equation}
\begin{cases}
X(x=0)=0 \\
X(x=L)=0
\end{cases}
\quad
\Rightarrow
\quad
\begin{cases}
cos(\phi)=0 \\
cos(k L + \phi)=0
\end{cases}
\quad
\Rightarrow
\quad
\begin{cases}
\phi=\frac{\pi}{2} + \mathbb{N} \pi \Rightarrow cos(kx+\phi)=sin(kx) \\
kL+\phi = \frac{\pi}{2}+\mathbb{N} \pi \Rightarrow kL=\mathbb{N} \pi
\end{cases}
\end{equation}

$k$ est le nombre d'onde lié à la période spatiale c'est à dire la longueur d'onde $\lambda$ par la relation $k=\frac{2 \pi}{\lambda}$. Finalement, une expression générale de la solution est donc obtenue par combinaison linéaire sur tous les modes.

\begin{equation}
u(x,t) = \sum_{n=1}^{\infty} u_n(x,t) = \sum_{n=1}^{\infty} X_n(x) T_n(t) \quad \text{avec} \quad u_n(x,t)=A_n sin \left( n \pi \frac{x}{L} \right) cos \left(  \frac{n c \pi}{L}t + \psi_n \right)
\end{equation}

C'est le principe de superposition des petits mouvements. Les coefficients $A_n$ et $\psi_n$ sont à déterminer à partir des conditions initiales sur le déplacement $u(x,t=0)$ et la vitesse $u_{,t}(x,t=0)$.

La solution permet de mettre en évidence les pulsations propres (ou fréquences propres) :

\begin{equation}
\omega_n=n\frac{\pi}{L}c=n \frac{\pi}{L}\sqrt{\frac{T_0}{\mu}} \quad \quad \left( f_n=\frac{\omega_n}{2 \pi}=\frac{n}{2L}\sqrt{\frac{T_0}{\mu}} \right)
\end{equation}

$n=1$ donne la fréquence fondamentale et $n \ge 2$ les harmoniques.

Pour simplifier, considérons un seul mode et réarrangeons la forme de la solution :

\begin{equation}
u(x,t) = A sin \left( kx \right) cos \left( \omega t + \psi \right) = \frac{A}{2} cos \left(\omega t - kx + \psi \right) + \frac{A}{2} cos \left(\omega t + kx + \psi \right)
\end{equation}

L'onde stationnaire peut donc s'écrire comme la superposition de 2 O.P.P. de même pulsation et de même amplitude se propageant en sens inverse.

### 0-d) Modes propres

Les modes propres sont donnés par la fonction $X_n(x)=sin \left( n \pi \frac{x}{L} \right)$. La solution est donc la combinaison linéaire de ces déformées vibrant à des fréquences multiples de la fondamentale (voir figure ci-dessous). Leur intensité et leur déphasage respectif est donné par les conditions initiales.

La déformée modale permet de mettre en évidence les noeuds et les ventres de vibrations :

Les noeuds $x_{nd}$ sont les points fixes relativement à un mode de rang $n$ donné :

\begin{equation}
\left. sin \left( n \pi \frac{x}{L} \right) \right\vert_{x=x_{nd}} = 0 \Rightarrow  n \pi \frac{x_{nd}}{L} = p \pi \Rightarrow x_{nd}= \frac{p}{n} L
\end{equation}

Avec $p \in \mathbb{N}$ et $0 \le p \le n$ sinon la position $x_{nd}$ ne serait pas localisée sur la corde.

Les ventres $x_{vt}$ sont les points les plus mobiles relativement à un mode de rang $n$ donné :

\begin{equation}
\left.  \frac{ d \left( sin \left( n \pi \frac{x_{vt}}{L} \right) \right)}{dx}\right\vert_{x=x_{vt}} = 0 \Rightarrow  n \pi \frac{x_{vt}}{L} = (2 p+1) \frac{\pi}{2} \Rightarrow x_{vt}= \frac{2p+1}{2n} L
\end{equation}

Avec $p \in \mathbb{N}$ et $0 \le p \le (2n-1)/2$ sinon la position $x_{vt}$ ne serait pas localisée sur la corde.

On peut mettre en évidence l'orthogonalité des modes dans l'espace de Hilbert, à partir du calcul du produit scalaire suivant :

\begin{equation}
\left\langle X_n(x),X_p(x) \right\rangle =\int_0^L X_n(x) \, X_p(x) dx = \int_0^x sin \left( n \pi \frac{x}{L} \right) sin \left( p \pi \frac{x}{L} \right) dx
\end{equation}

On procède à deux Intégrations Par Parties successives :

\begin{equation}
\begin{array}{l c l}
\left\langle X_n(x),X_p(x) \right\rangle & = & \int_0^x sin \left( n \pi \frac{x}{L} \right) sin \left( p \pi \frac{x}{L} \right) dx \\ & = & -\frac{L}{n \pi} \left[ cos \left(n \pi \frac{x}{L} \right) sin \left(p \pi \frac{x}{L} \right) \right]_0^L+ \frac{n}{p} \int_0^x cos \left( n \pi \frac{x}{L} \right) cos \left( p \pi \frac{x}{L} \right) dx \\ & = & \frac{n}{p} \left( \frac{L}{n \pi} \left[ sin \left(n \pi \frac{x}{L} \right) cos \left(p \pi \frac{x}{L} \right) \right]_0^L+ \frac{n}{p} \int_0^x sin \left( n \pi \frac{x}{L} \right) sin \left( p \pi \frac{x}{L} \right) dx \right) \\ & = & \left(  \frac{n}{p} \right)^2 <X_n(x),X_p(x)>
\end{array}
\end{equation}

Ainsi, si $n=p$, $\left\langle X_n(x),X_p(x) \right\rangle$ peut prendre n'importe quelle valeur. Si $n \neq p$, $\left\langle X_n(x),X_p(x) \right\rangle$ vaut forcément 0.

```python
plt.rcParams['figure.figsize'] = [12, 3]
xsL=np.linspace(0,1,100)
for n in range(1,5):
    plt.plot(xsL,np.sin(n*np.pi*xsL),label=str(n))
    for p in range(0,n+1):
        plt.plot([p/n,p/n],[-1,1],'k:')
        if p<=(2*n-1)/2:
            plt.plot([(2*p+1)/2/n,(2*p+1)/2/n],[-1,1],'c--')
plt.xlabel("$x/L$")
plt.ylabel("$g_n(x)$")
plt.title("Modes propres de la corde vibrante fixée à ses extrémités")
plt.legend()
plt.grid()
```

### 0-d) Résonance

<p align="center">
  <img src="corde_melde.png" width="400"/>
</p>

Pour mettre en évidence le phénomène de résonance, on utilise le dispositif de la corde de Melde en régime forcé.
L'extrémité gauche de la corde est excitée à une pulsation $\omega_{imp}$, tandis que l'extrémité droite est fixée. Les conditions aux limites s'écrivent alors :

\begin{equation}
\begin{cases}
u(x=0,t)=a \cos{\omega_{imp} t} \\
u(x=L,t)=0
\end{cases}
\end{equation}

On repart de la forme de solution $u_n(x,t) = A_n cos \left(k_n x+ \Phi_n \right) cos \left( \omega_n t + \Psi_n \right)$.

En $x=L$, on a :

\begin{equation}
X_n(x=L)=cos(k_n L + \Phi_n)=0 \Rightarrow k_n L+\Phi_n=\frac{\pi}{2}+\mathbb{N}\pi \Rightarrow cos \left(k_n x + \Phi_n \right)=sin \left( k_n (L-x) \right)
\end{equation}

En $x=0$, on a :

\begin{equation}
u_n(0,t)=A sin (k_nL) cos \left(\omega_n t+\Psi_n \right) = a cos \left( \omega_{imp} t \right)
\quad
\Rightarrow
\quad
\begin{cases}
\omega_n=\omega_{imp} \\
\Psi_n=0 \\
A sin(k_n L)=a
\end{cases}
\end{equation}

Finalement :

\begin{equation}
u(x,t) = \sum_{n=1}^{\infty} u_n(x,t) = \sum_{n=1}^{\infty} X_n(x) T_n(t) \quad \text{avec} \quad u_n(x,t)=\frac{a}{sin (k_n L)} cos (\omega_{imp} t) sin\left( k_n(L-x) \right)
\end{equation}

Puisque $k_n=\omega_n /c$ et $k_n L= n\pi$, exciter le système à une pulsation propre ($\omega_{imp}=\omega_n$) revient à faire tendre le coefficient $\frac{a}{sin (k_n L)}$ vers l'infini (amortissement négligé ici). La contribution des autres modes devient donc négligeable. On retrouve (au signe près) la déformée déterminée précédemment, puisque :

\begin{equation}
sin\left( k_n(L-x) \right)= \pm sin\left( n \pi \frac{x}{L} \right)
\end{equation}

### 0-e) Application 1 : cas de la corde frappée (piano)

Rappel sur la décomposition de Fourier :

Une fonction périodique f(t) de période T peut, sous certaines conditions mathématiques qui seront toujours réalisées
dans la pratique en physique, se décomposer en une somme de fonctions sinusoïdales de la forme :

\begin{equation}
f(t)=a_0+\sum_{n=1}^{\infty}\left( a_n \cos{n \omega t} + b_n \sin{n \omega t} \right) \quad \text{avec} \quad \begin{cases}
a_0=\frac{1}{T} \int_0^T f(t) dt \\
a_n=\frac{2}{T} \int_0^T f(t) \cos{n \omega t} dt \\
b_n=\frac{2}{T} \int_0^T f(t) \sin{n \omega t} dt \\
\end{cases}
\end{equation}

<p align="center">
  <img src="corde_frappe.png" width="400"/>
</p>

On cherche maintenant à déterminer la solution complète, tenant compte des conditions initiales. On s'intéresse au cas de la corde de piano, qui est frappée à l'instant initial à l'emplacement $x \in [a,a+e]$ par un maillet de dimension faible face à la longueur de la corde :

\begin{equation}
\begin{cases}
u(x,0)=0 \quad \forall x \\
\frac{\partial u}{\partial t}(x,0)= \begin{cases}
v_0 \quad \forall x \in [a, a+e] \\
0 \quad \forall x \notin [a, a+e] 
\end{cases}
\end{cases}
\end{equation}

On repart de la solution tenant compte des fixations aux extrémités, tout en modifiant l'expression en vue d'une décomposition en séries de Fourier :

\begin{equation}
\begin{cases}
u(x,t) = \sum_{n=1}^{\infty} A_n sin \left( n \pi \frac{x}{L} \right) cos \left(  \frac{n c \pi}{L}t + \Psi_n \right)  = \sum_{n=1}^{\infty} sin \left( n \pi \frac{x}{L} \right) \left( a_n cos \left( \frac{n c \pi}{L}t \right) + b_n sin \left( \frac{n c \pi}{L}t \right) \right) \\
u_{,t}(x,t) = \sum_{n=1}^{\infty} sin \left( n \pi \frac{x}{L} \right) \frac{n c \pi}{L} \left( - a_n sin \left( \frac{n c \pi}{L}t \right) + b_n cos \left( \frac{n c \pi}{L}t \right) \right)
\end{cases}
\end{equation}

On applique les conditions initiales, en commençant par le déplacement initial :

\begin{equation}
u(x,0)=\sum_{n=1}^\infty a_n sin \left( n \pi \frac{x}{L} \right)=\sum_{n=1}^\infty a_n sin \left( n 2\pi \frac{x}{2L} \right)
\end{equation}

On reconnait une décomposition en séries de Fourier de période $2L$. On rend la fonction périodique pour utiliser la décomposition en série de Fourier. Il vient :

\begin{equation}
a_n= \frac{2}{2L} \int_{-L}^{+L} u(x,0) sin \left( n 2 \pi \frac{x}{2L} \right) dx
\end{equation}

Puisque $u(x,0)=0$, $a_n=0$ quelque soit $n$.

On considère ensuite la vitesse initiale :

\begin{equation}
u_{,t}(x,0) = \sum_{n=1}^{\infty} b_n \frac{n c \pi}{L} sin \left( n \pi \frac{x}{L} \right)
\end{equation}

Ainsi, en exploitant la symétrie de l'intégrande,

\begin{equation}
b_n^* = b_n \frac{n \pi c}{L} = \frac{1}{L} \int_{-L}^{+L} u_{,t}(x,0) sin \left( n \pi \frac{x}{L} \right) dx= \frac{2}{L} \int_{0}^{+L} u_{,t}(x,0) sin \left( n \pi \frac{x}{L} \right) dx
\end{equation}

Le calcul de l'intégrale nous amène à :

\begin{equation}
b_n=\frac{2}{n \pi c} \int_{a}^{a+e} v_0 sin \left( n \pi \frac{x}{L} \right) dx \approx \frac{2}{n \pi c} v_0 sin \left( n \pi \frac{a}{L} \right) e
\end{equation}

Il vient :

\begin{equation}
u(x,t) = \sum_{n=1}^{\infty} sin \left( n \pi \frac{x}{L} \right) \frac{2}{n \pi c} v_0 sin \left( n \pi \frac{a}{L} \right) e \, sin \left( \frac{n c \pi}{L}t \right)
\end{equation}

```python
plt.rcParams['figure.figsize'] = [12, 3]
n=np.arange(1,10)
bn=1/n
plt.stem(bn,'r')
plt.xlabel("$n$")
plt.ylabel("$b_n/max(b_n)$")
plt.title("Contribution normalisée des modes")
plt.grid()
```

### 0-f) Application 2 : cas de la corde pincée (guitare)

<p align="center">
  <img src="corde_pince.png" width="400"/>
</p>

On s'intéresse au cas de la corde de guitare initialement déviée selon un profil triangulaire de sommet correspondant à $x=a$ :

\begin{equation}
\begin{cases}
u(x,0)= \begin{cases}
h \frac{x}{a} \quad \forall x \in [0, a] \\
h \frac{L-x}{L-a} \quad \forall x \in [a, L] 
\end{cases} \\
\frac{\partial u}{\partial t}(x,0)=0 \quad \forall x
\end{cases}
\end{equation}

De même que précédemment :

\begin{equation}
a_n= \frac{1}{L} \int_{-L}^{+L} u(x,0) sin \left( n \pi \frac{x}{L} \right) dx= \frac{2}{L} \int_{0}^{+L} u(x,0) sin \left( n \pi \frac{x}{L} \right) dx
\end{equation}

Et,

\begin{equation}
b_n^* = b_n \frac{n \pi c}{L} = \frac{1}{L} \int_{-L}^{+L} u_{,t}(x,0) sin \left( n \pi \frac{x}{L} \right) dx
\end{equation}

Puisque la vitesse est initialement nulle en tout point. Ainsi, $b_n=0$.

Enfin, 

\begin{equation}
a_n = \frac{2}{L} \left( \int_{0}^{a} u(x,0) sin \left( n \pi \frac{x}{L} \right) + \int_{a}^{h} u(x,0) sin \left( n \pi \frac{x}{L} \right) \right) dx=\frac{2h}{n^2\pi^2} \frac{1}{\alpha (1-\alpha)} sin \left( n \pi \alpha \right) \quad \text{avec} \quad \alpha=\frac{a}{L}
\end{equation}

```python
plt.rcParams['figure.figsize'] = [12, 3]
n=np.arange(1,10)
an=1/n**2
plt.stem(bn,'r',label="piano")
plt.stem(an,'g--',label="guitare")
plt.xlabel("$n$")
plt.ylabel("$a_n/max(a_n)$")
plt.title("Contribution normalisée des modes")
plt.legend()
plt.grid()
```

### 0-g) Equation de d'Alembert avec prise en compte du poids propre

On repart des mêmes hypothèses mais cette fois, la corde est soumise à un champ de pesanteur dirigé selon la verticale descendante.

\begin{equation}
\mu \, dx \, u_{,tt} (x,t) \overrightarrow{y} = \overrightarrow{T}(x+dx,t)-\overrightarrow{T}(x,t) + \mu \, dx \, g \overrightarrow{y}
\end{equation}

La projection sur $\overrightarrow{x}$ est inchangée. On projette sur $\overrightarrow{y}$ :
\begin{equation}
\mu u_{,tt} \approx T_0 u_{,xx} - \mu g
\end{equation}

Il vient l'équation de d'Alembert :

\begin{equation}
c^2 u_{,xx}-u_{,tt}=g
\end{equation}

A la solution générale déterminée précédemment, s'ajoute la solution particulière (permanente et indépendante de $t$) satisfaisant :

\begin{equation}
u_{p,xx}=\frac{g}{c^2} \Rightarrow u_p(x)= \frac{g}{2 c^2} x^2 + Bx + C
\end{equation}

Les CL $y(0,t)=y(L,t)=0$ imposent $C=0$ et $B=-\frac{gL}{2c^2}$. Finalement, $u_{p}(x)=-\frac{g}{2 c^2} x(L-x)$.

On obtient une déformée parabolique qui n'affecte en rien la réponse dynamique de la corde.

```python
plt.plot(xsL,-xsL*(1-xsL)/max(xsL*(1-xsL)),'r.')
plt.xlabel("$x/L$")
plt.ylabel("$u/\| u_{max} \|$")
plt.title("Effet du poids propre")
plt.grid()
```

### 0-h) Equation de d'Alembert avec frottements visqueux

On repart des mêmes hypothèses (sans la pesanteur) et on ajoute une force de frottement fluide distribuée le long de la corde $d\overrightarrow{f}=-\gamma u_{,t}(x,t)$.

La projection du PFD selon $\overrightarrow{x}$ puis $\overrightarrow{y}$ donne l'équation :

$$
u_{,tt} + \frac{\gamma}{\mu} u_{,t} = c^2 u_{,xx}
$$

On suppose toujours que la solution est de la forme $u(x,t)=X(x) T(t)$. L'équation de d'Alembert devient :

\begin{equation}
c^2 X''(x) T(t)= X(x) \ddot{T}(t) + \frac{\gamma}{\mu} X(x) \dot{T}(t) \Rightarrow c^2 \frac{X''(x)}{X(x)}=\frac{\ddot{T}(t)}{T(t)}+\frac{\gamma}{\mu} \frac{\dot{T}(t)}{T(t)}=-\omega^2
\end{equation}

Avec $-\omega^2$. Il vient :

\begin{equation}
\begin{cases}
X''(x)+k^2 X(x)=0 \\
\ddot{T}(t)+\frac{\gamma}{\mu} \dot{T}(t)+\omega^2 T(t)=0
\end{cases}
\quad
\Rightarrow
\quad
\begin{cases}
X(x)=A \, cos(kx+\phi) \\
T(t)=B \, e^{-\frac{\gamma}{2\mu}t} cos(\omega^* t + \psi)
\end{cases}
\end{equation}

Avec $\omega^*=\omega \sqrt{1-\left( \frac{\gamma}{2 \mu \omega} \right)^2}$. Finalement,

\begin{equation}
u(x,t)=K \, cos(kx+\phi) e^{-\frac{\gamma}{2 \mu}t} cos(\omega^* t + \psi)
\end{equation}

```python
plt.rcParams['figure.figsize'] = [12, 3]
t=np.linspace(0,100,10000)
f=0.5
mu=1
gamma1=0.05
gamma2=0.1
omega=2*np.pi*f
xi1=gamma1/2/omega/mu
xi2=gamma2/2/omega/mu
omega_r1=omega*np.sqrt(1-xi1**2)
omega_r2=omega*np.sqrt(1-xi2**2)
u=np.cos(omega*t)
u_amort1=np.exp(-xi1*omega*t)*np.cos(omega_r1*t)
u_amort2=np.exp(-xi2*omega*t)*np.cos(omega_r2*t)
plt.plot(t,u,'r')
plt.plot(t,u_amort1,'g')
plt.plot(t,u_amort2,'b')
plt.xlabel("t")
plt.ylabel("u")
plt.legend(["$\gamma=0$","$\gamma=0.05$","$\gamma=0.1$"])
```

### 0-i) Approche énergétique

On ne tient pas compte de la pesanteur, ni des frottements, et on reprend les hypothèses précédentes. La portion de corde de masse $dm=\mu dx$ et de vitesse $u_{,t}$ a pour énergie cinétique $dEc=\frac{1}{2} \mu dx u_{,t}^2$. La densité linéique associée est donc $e_c=\frac{dEc}{dx} = \frac{1}{2} \mu u_{,t}^2$.

Pour déterminer l'énergie potentielle, il faut passer par le travail correspondant à l'allongement sous l'effet de la tension supposée constante $W=T_0 \Delta L$.

L'allongement se calcul ainsi :

\begin{equation}
\Delta L = l-L = \int_0^L ds - L = \int_0^L \sqrt{1+u_{,x}^2} dx - L \approx \int_0^L \left( 1 + \frac{1}{2} u_{,x}^2 \right) dx - L = \frac{1}{2} \int_0^L u_{,x}^2 dx
\end{equation}

L'énergie potentielle de la corde est donc donnée par $dE_p=\frac{T_0}{2} \int_0^L u_{,x}^2 dx$. La densité linéique associée est donc $e_p=\frac{dEp}{dx} = \frac{T}{2} u_{,x}^2$.

On somme les densités d'énergie cinétique et potentielle $e=e_c+e_p=\frac{\mu}{2} u_{,t}^2 + \frac{T_0}{2} u_{,x}^2$, puis on dérive.

\begin{equation}
\frac{\partial e}{\partial t}= \mu u_{,t} u_{,tt} + T_0 u_{,x} y_{,xt} = \frac{T}{c^2} u_{,t} u_{,tt} + T_0 \left( \frac{\partial}{\partial x} \left( u_{,x} u_{,t} \right) - u_{,xx} u_{,t} \right) = - T_0 u_{,t} \left( u_{,xx} - \frac{1}{c^2} u_{,tt} \right) + \frac{\partial}{\partial x} \left( T_0 u_{,x} u_{,t} \right)=\frac{\partial}{\partial x} \left( T_0 u_{,x} u_{,t} \right)
\end{equation}

Dans le dernier membre de l'équation, on voit apparaître l'équation de d'Alembert (en absence de gravité). Finalement, $\frac{\partial e}{\partial t}+\frac{\partial S}{\partial x} = 0$, avec la puissance $S= - T_0 y_{,x} y_{,t}$, ce qui correspond  à l'équation locale de la conservation d'énergie.

A l'inverse, une corde tendue entre deux points fixes impose les puissances entrante et sortante nulles ($S(0)=S(L)=0$) ie que l'énergie mécanique soit constante $E= \int e dx = cste$. En effet,

$$
\frac{dE}{dt}=\int \frac{\partial e}{\partial t} dx = - \int_0^L \frac{\partial S}{\partial x} dx = \left[ T u_{,x} u_{,t} \right]_0^l=0=S(0)-S(L)
$$

En repartant de $\frac{\partial S}{\partial x}+\frac{\partial e}{\partial t}=-T_0 \frac{\partial u}{\partial t} \left( u_{,xx}-\frac{1}{c^2} u_{,tt} \right)$ on retrouve l'équation de d'Alembert.

### 0-j) Principe de moindre action

On ne tient pas compte de la pesanteur, ni des frottements, et on reprend les hypothèses précédentes. L'énergie cinétique d'une portion de corde s'écrit toujours $dEc=\frac{1}{2} \mu dx u_{,t}^2$, et l'énergie potentielle $dE_p=\frac{T_0}{2} u_{,x}^2 dx$.

On peut donc écrire le Lagrangien élémentaire suivant :

\begin{equation}
d\mathcal{L}=\frac{1}{2} \left( \mu \left( \frac{\partial y}{\partial t} \right)^2 - T_0 u_{,x}^2 \right) dx
\end{equation}

Le Lagrangien de la corde est la somme des Lagrangiens élémentaires :

\begin{equation}
\mathcal{L} = \int d \mathcal{L} = \frac{1}{2} \int_0^L \left( \mu u_{,t}^2 - T_0 u_{,x}^2 \right) dx
\end{equation}

L'action de la corde au sens de Hamilton s'écrit :

\begin{equation}
\mathcal{S} = \frac{1}{2} \int_t \int_0^L \left( \mu \left( u_{,t} \right)^2 - T_0 u_{,x}^2 \right) dx dt
\end{equation}

Le principe de moindre action affirme qu'il faut minimiser l'intégrande $I=\left( \mu u_{,t}^2 - T_0 u_{,x}^2 dx \right)$, ce qui équivaut à résoudre l'équation d'Euler-Lagrange :

\begin{equation}
I_{,u}=\frac{\partial}{\partial t} \left( \frac{\partial I}{\partial \left( u_{,t} \right)} \right) + \frac{\partial}{\partial x} \left( \frac{\partial I}{\partial \left( u_{,x} \right)} \right)
\end{equation}

Or $I_{,u}=0$. Il vient donc :

\begin{equation}
0=\frac{\partial}{\partial t} \left( 2 \mu u_{,t} \right) + \frac{\partial}{\partial x} \left( 2T_0 u_{,x} \right)
\end{equation}

Ce qui permet de retrouver l'équation de d'Alembert.

## 1-Généralités

### 1-a) Première classification des EDP

Equations aux dérivées partielles (EDP) linéaires du second ordre à deux variables indépendantes :

\begin{equation}
A u_{,xx}+ B u_{,xy} + C u_{,yy} + D u_{,x} + E u_{,y} + F u + G =0 \quad (1)
\end{equation}
Où les coefficients A, ..., G sont constants.

On classe les EDP selon la classification des coniques. Ainsi :

i) EDP elliptiques  : $B^2 - 4AC<0$, ii) EDP paraboliques  : $B^2 - 4AC=0$, iii) EDP hyperboliques  : $B^2 - 4AC>0$.

```python
fig, ax = plt.subplots(figsize=(5,5))

# paramètre
t = np.linspace(-3,3,400)

# --- Ellipse : x^2/4 + y^2 = 1  -> 4x^2 + y^2 -4 =0
theta = np.linspace(0,2*np.pi,400)
x_e = 2*np.cos(theta)
y_e = np.sin(theta)
a,b,c = 1,0,4
delta = b**2 - 4*a*c
ax.plot(x_e,y_e,color="magenta",label=f"Ellipse  (Δ={delta})")

# --- Parabole : y = x^2 -> x^2 - y =0
x_p = t
y_p = t**2
a,b,c = 1,0,0
delta = b**2 - 4*a*c
ax.plot(x_p,y_p,color="green",label=f"Parabole  (Δ={delta})")

# --- Hyperbole : x^2 - y^2 = 1
u = np.linspace(-2,2,400)
x_h = np.cosh(u)
y_h = np.sinh(u)
a,b,c = 1,0,-1
delta = b**2 - 4*a*c
ax.plot(x_h,y_h,color="blue",label=f"Hyperbole  (Δ={delta})")
ax.plot(-x_h,y_h,color="blue")  # seconde branche

ax.set_title("ax² + bxy + cy² + dx + ey + f = 0")
ax.set_xlabel("x")
ax.set_ylabel("y")
#ax.set_aspect("equal")
ax.grid(True)
ax.legend()

plt.show()
```

### 1-b) Conditions auxiliaires

On dit qu'une EDP et un jet de conditions auxiliaires (Conditions Initiales ou aux Limites) est bien posé au sens d'Hadamard si :

i) la solution existe, ii) la solution est unique, iii) la solution dépend continûment des données auxiliaires.

Les conditions auxiliaires (initiales et aux limites) sont couramment classées sous 3 formes :

i) les conditions de Dirichlet ($u=f$ sur $\partial R$), ii) les conditions de von Neumann ($u_{,\boldsymbol{n}}=\boldsymbol{f}$ ou $u_{\boldsymbol{s}}=\boldsymbol{g}$ sur $\partial R$), iii) les conditions de Robin ($u_{,\boldsymbol{n}}+\boldsymbol{k}u=\boldsymbol{f}$). Avec $\partial R$ la frontière du domaine considéré.

### 1-c) Les courbes caractéristiques

Lorsque les EDP sont fonctions de 2 variables indépendantes, la classification peut être obtenue en cherchant des directions caractéristiques selon lesquelles les équations n'impliquent plus que des différentielles totales.

(1) peut s'écrire de façon réduite :

\begin{equation}
A u_{,xx}+ B u_{,xy} + C u_{,yy} + H =0 \quad (2)
\end{equation}

On pose $P=u_{,x}$, $Q=u_{,y}$, $R=u_{,xx}$, $S=u_{,xy}$ et $T=u_{,yy}$.

Soit une courbe $\Gamma$ à l'intérieur du domaine, où $P$, $Q$, $R$, $S$ et $T$ vérifient (2). $\frac{dy}{dx}$ définit la pente de la tangente à $\Gamma$.

\begin{equation}
\begin{cases}
du_{,x}=dP=u_{,xx} dx + u_{,xy} dy = R dx + S dy \\
du_{,y}=dQ=u_{,xy} dx + u_{,yy} dy = S dx + T dy \\
\end{cases}
\quad
\Rightarrow
\quad
AR+BS+CT+H=A \left(\frac{dP}{dx}-S \frac{dy}{dx} \right)+BS+C\left( \frac{dQ}{dy}-S \frac{dx}{dy} \right) + H=0
\end{equation}

(2) s'écrit finalement :

\begin{equation}
S \left( A \left( \frac{dy}{dx} \right)^2 - B \frac{dy}{dx} + C \right) - \left( A \frac{dP}{dx} + H \right) \frac{dy}{dx} - C \frac{dQ}{dx} = 0
\end{equation}

En choisissant $\frac{dy}{dx}$ tel que $A \left( \frac{dy}{dx} \right)^2 - B \frac{dy}{dx} + C = 0$ (résolution d'une équation du second degré), (2) se réduit à une simple relation entre $\frac{dP}{dx}$ et $\frac{dQ}{dx}$ :

\begin{equation}
\left( A \frac{dP}{dx} + H \right) \frac{dy}{dx} + C \frac{dQ}{dx} = 0 \quad (3)
\end{equation}

### 1-d) Deuxième classification des EDP

La résolution de $A \left( \frac{dy}{dx} \right)^2 - B \frac{dy}{dx} + C=0$ définit les directions des caractéristiques (dans $\mathbb{C}$), le long desquelles (3) est vérifiée. On peut ainsi donner une nouvelle classification des EDP :

i) EDP elliptiques  : $B^2 - 4AC<0$ $\Rightarrow$ il existe deux caractéristiques complexes.

ii) EDP paraboliques  : $B^2 - 4AC=0$ $\Rightarrow$ il existe une caractéristique réelle.

iii) EDP hyperboliques  : $B^2 - 4AC>0$ $\Rightarrow$ il existe deux caractéristiques réelles.

## 2-Equation des cordes vibrantes

On rappelle l'équation d'une corde vibrante (pour une corde de longueur $L$, de densité linéique de masse $\mu$ et tendue par une force $T_0$) donnée par l'équation de d'Alembert :

\begin{equation}
u_{,tt} - c^{2} u_{,xx}=0
\end{equation}

Où $u=u(x,t)$ représente le déplacement transversal en tout instant et tout point de la corde et $c=\sqrt{\frac{T_0}{\mu}}$ la vitesse de propagation de l'onde transversale.

On a dans ce cas $A=1$ et $C=-c^{2}$. Ainsi $B^2 - 4AC= 4c^2 >0$. Il s'agit donc d'une EDP hyperbolique.

Les directions caractéristiques sont données par $\frac{dx}{dt}$, telles que $A \left( \frac{dx}{dt} \right)^2 - B \frac{dx}{dt} + C = \left( \frac{dx}{dt} \right)^2 - c^2 = 0$, soit :

\begin{equation}
\frac{dx}{dt}=\pm c
\end{equation}

Les caractéristiques passant par un point $P(x_i=0.5,t_j=0.5)$ sont données ci-dessous :

```python
plt.rcParams['figure.figsize'] = [4, 4]
c=1
x=np.linspace(0,1,100)
plt.plot(x,x/c,'b',label="caractéristique $\Gamma^+$")
plt.plot(x,1-x/c,'r',label="caractéristique $\Gamma^-$")
plt.plot(0.5,0.5,'ko',label="P(0.5,0.5)")
plt.legend()
plt.xlabel("t")
plt.ylabel("x")
```

La relation $(3)$ se réduit pour le cas de la corde vibrante à :

$$
\lambda \frac{du_{,t}}{dt}-c^2 \frac{du_{,x}}{dt}=\frac{d \left(\lambda u_{,t}-c^2 u_{,x} \right)}{dt}=0
$$

Avec $\frac{d}{dt}=\frac{\partial}{\partial t}+\frac{dx}{dt} \frac{\partial}{\partial x}$ et $\lambda = \pm c$.

Il vient $\frac{d \left(u_{,t} \pm c u_{,x} \right)}{dt}=0$, soit :

$$
u_{,t} \pm c u_{,x}=constante
$$

Ce qui permet de retrouver le fait que l'onde se propage dans les deux sens de l'espace.

```python
def calculer_frequence(L, tension, mu):
    """Calcule la fréquence fondamentale f = 1/(2L) * sqrt(T/mu)"""
    return (1 / (2 * L)) * np.sqrt(tension / mu)

# --- Paramètres par défaut (Corde de La d'une guitare) ---
L_standard = 0.715      # 65 cm
mu_standard = 0.004     # 4 g/m soit 0.004 kg/m
tension_standard = 85 # Newtons (environ 7kg de force)

# 1. Variation de la Tension (L'action d'accorder)
tensions = np.linspace(10, 150, 100)
freq_vs_T = calculer_frequence(L_standard, tensions, mu_standard)

# 2. Variation de la Distance (Appuyer sur une frette)
longueurs = np.linspace(0.1, 0.715, 100)
freq_vs_L = calculer_frequence(longueurs, tension_standard, mu_standard)

# --- Visualisation ---
plt.figure(figsize=(14, 5))

# Graphique de l'accordage (Tension)
plt.subplot(1, 2, 1)
plt.plot(tensions, freq_vs_T, color='green', lw=2)
plt.axhline(y=82, color='black', linestyle='--', label="Note MI (82 Hz)")
plt.title("L'Accordage : Fréquence vs Tension")
plt.xlabel("Tension T (Newtons)")
plt.ylabel("Fréquence f (Hz)")
plt.legend()
plt.grid(True, alpha=0.3)

# Graphique de la mélodie (Longueur)
plt.subplot(1, 2, 2)
plt.plot(longueurs, freq_vs_L, color='blue', lw=2)
plt.title("La Mélodie : Fréquence vs Longueur")
plt.xlabel("Longueur vibrante L (m)")
plt.ylabel("Fréquence f (Hz)")
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# --- Petit bonus : Calcul d'accordage précis ---
f_cible = 82 # Fréquence du mi
t_requise = mu_standard * (2 * L_standard * f_cible)**2
print(f"Pour obtenir un La ({f_cible} Hz) avec L={L_standard}m et mu={mu_standard}kg/m :")
print(f"La tension doit être réglée à : {t_requise:.2f} Newtons")
```

```python
import numpy as np
import scipy.io.wavfile as wav

# --- Paramètres de l'instrument ---
L_diapason = 715  # Longueur de la corde vibrante en mm (standard guitare)
fe = 44100          # Fréquence d'échantillonnage
duree_note = 0.8
alpha = 0.15        # Position du pincement (main droite)
n_harmo = 20        # Richesse du son

def distance_frette(n, L):
    """Calcule la distance de la frette n par rapport au sillet (en mm)"""
    return L * (1 - 2**(-n/12))

def generer_note(f0, duree, fs, a, n_h):
    t = np.linspace(0, duree, int(fs * duree), endpoint=False)
    signal = np.zeros_like(t)
    for n in range(1, n_h + 1):
        # Formule de Fourier de ton slide
        amplitude = np.sin(n * np.pi * a) / (n**2 * a * (1 - a)*np.pi**2)
        enveloppe = np.exp(-t * (1 + n*0.3)) # Décroissance naturelle
        signal += amplitude * np.sin(2 * np.pi * (n * f0) * t) * enveloppe
    return signal / np.max(np.abs(signal))

# --- Initialisation ---
f_mi_grave = 82.41
noms = ["Mi", "Fa", "Fa#", "Sol", "Sol#", "La", "La#", "Si", "Do", "Do#", "Ré", "Ré#", "Mi octave"]
audio_final = []

print(f"{'Note':<12} | {'Fréquence (Hz)':<15} | {'Distance Frette (mm)':<20}")
print("-" * 55)

for i in range(13):
    # 1. Calcul de la fréquence
    freq = f_mi_grave * (2**(i/12))
    
    # 2. Calcul de la position de la frette
    dist = distance_frette(i, L_diapason)
    
    print(f"{noms[i]:<12} | {freq:<15.2f} | {dist:<20.2f}")
    
    # 3. Génération du son
    note = generer_note(freq, duree_note, fe, alpha, n_h=n_harmo)
    audio_final.append(note)
    audio_final.append(np.zeros(int(fe * 0.05)))

# Exportation
audio_concat = (np.concatenate(audio_final) * 32767).astype(np.int16)
wav.write("gamme_complete.wav", fe, audio_concat)

print("\nFichier 'gamme_complete.wav' généré.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculer_spectre_theorique(f1, alpha, n_max=20):
    """
    Calcule la TF théorique basée sur l'équation an fournie.
    f1 : fréquence fondamentale (ex: 82.41 Hz)
    alpha : position relative du pincement (a/L)
    n_max : nombre d'harmoniques à calculer
    """
    indices_n = np.arange(1, n_max + 1)
    
    # Ta formule : an = (2h / (n^2 * pi^2)) * (1 / (alpha * (1 - alpha))) * sin(n * pi * alpha)
    # Note : On normalise 2h/pi^2 à 1 pour l'affichage relatif
    amplitudes = (1 / (indices_n**2 * alpha * (1 - alpha))) * np.sin(indices_n * np.pi * alpha)
    
    # Les fréquences des pics
    frequences = indices_n * f1
    
    return frequences, np.abs(amplitudes)

# --- Paramètres ---
f_mi = 82.41
alpha_pincement = 0.15 # Essayez 0.5 pour voir les harmoniques paires s'annuler

freqs, amps = calculer_spectre_theorique(f_mi, alpha_pincement)

# --- Affichage de la "Transformée de Fourier" théorique ---
plt.figure(figsize=(15, 10))

# On trace des barres (pics) pour représenter la TF discrète
plt.stem(freqs, amps, linefmt='b-', markerfmt='bo', basefmt=' ')

# Habillage
plt.title(f"Spectre théorique (TF) d'une corde pincée à $\\alpha$ = {alpha_pincement}")
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude $|a_n|$")
plt.xticks([f_mi * n for n in range(1, 11)]) # Marqueurs pour les 10 premières harmo
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(0, f_mi * 12)

# Ajout de labels sur les pics
for f, a, n in zip(freqs[:5], amps[:5], range(1, 6)):
    plt.text(f, a + 0.05, f"n={n}", ha='center')

plt.show()
```
